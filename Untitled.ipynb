```python
import os
import datetime
import pandas as pd
import oracledb
from dotenv import load_dotenv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
plt.rc(
    "figure",
    autolayout=True,
    figsize=(15, 8),
    titlesize=18,
    titleweight='bold',
)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=16,
    titlepad=10,
)
```

```python
load_dotenv()

USER = os.environ["DB_USER"]
PASSWORD = os.environ["DB_PASSWORD"]
CONFIG_DIR = os.environ["CONFIG_DIR"]
DSN = os.environ["DSN"]
```

```python
def query(query: str) -> pd.DataFrame:
    """
    This function connects to an oracle database and executes the query string
    and then closes the database connection.

    The result of the query is a `pandas.DataFrame` object
    """

    try:
        conn = oracledb.connect(user=USER, password=PASSWORD, dsn=DSN, config_dir=CONFIG_DIR)
        print("Connection OK!")
        cursor = conn.cursor()
        result = cursor.execute(query).fetchall()
        cols = [description[0] for description in cursor.description]
        df = pd.DataFrame(result, columns = cols)
        return df
        
    except Exception as e:
        print("Connection Failed")
        raise(e)
    
    finally:
        cursor.close()
        conn.close()
```

The transactions selected to compare volumes are `NIP`, `NAPS`, `NEFT` `Quickteller`, `Remita`, `Internet Banking Transfer`. These transactions have the following trn_code respectively `113, 194, 119, 118, 777, 132`

```python
transactions = query("""
select t.trn_desc, a.* from fcjlive.acvw_all_ac_entries a
join fcjlive.sttm_trn_code t on a.trn_code = t.trn_code
where a.trn_code in ('113', '194', '119', '118', '777', '132') 
and a.trn_dt between '1/jan/2022' and '31/dec/2023'
""")
```

```python
transactions.info()
```

```python
transactions.columns = transactions.columns.str.lower()
```

## Top Transactions By Volume

```python
fig, ax = plt.subplots()
transactions.trn_desc.value_counts().sort_values().plot.barh(color="#004987")
ax.bar_label(ax.containers[0], fmt="{:,.0f}")
plt.title("Transaction Type By Volume")
plt.ylabel("Transaction Type")
plt.xlabel("Volume")
plt.show()
```

## Transaction Volume Over Time

```python
transactions.sort_values("trn_dt").groupby("trn_dt").count()["trn_desc"].plot(color="#004987")
plt.xlabel("Transaction Date")
plt.show()
```

```python
transactions.sort_values("trn_dt").groupby("trn_dt").count()["trn_desc"].rolling(30).mean().plot(color="#004987")
plt.xlabel("Transaction Date")
plt.text(x=datetime.date(2022, 5, 1), y=3200, size=15,
         s="Average Daily Transactions Per Month Shows A Growing Trend Between January 2022 and December 2023")
plt.show()
```

```python
def get_transactions(trn_code: str) -> pd.DataFrame: 
    return (transactions
            .query(f"trn_code == '{trn_code}'")
            .sort_values("trn_dt")
            .reset_index(drop=True)
           )
```

```python
def plot_transaction_volume(df: pd.DataFrame):
    return (df.groupby("trn_dt")
            .agg({"trn_desc": "count"})
            .rename({"trn_desc": "volume"}, axis=1)
            .plot(color="#004987")
           )


def plot_transaction_volume_rolling(df: pd.DataFrame, period: int):
    return (df
            .groupby("trn_dt")
            .agg({"trn_desc": "count"})
            .rename({"trn_desc": "volume"}, axis=1)
            .rolling(period).mean()
            .dropna()
            .plot(color="#004987")
           )
    
```

## Quickteller Transaction Volume

```python
quickteller = get_transactions("118")
```

```python
plot_transaction_volume(quickteller)
plt.xlabel("Transaction Date")
plt.title("Quickteller Transactions 2022 - 2023")
plt.show()
```

```python
plot_transaction_volume_rolling(quickteller, period=30)
plt.xlabel("Transaction Date")
plt.text(x=datetime.date(2022, 2, 1), y=250, size=15,
         s="Quickteller Transactions Daily Average Peaked in April of 2023 Before Dropping and Then Reaching Its Lowest in December of 2023")
plt.show()
```

Quickteller transactions peaked in April of 2023 before dropping and then Reaching its lowest in December of 2023.
