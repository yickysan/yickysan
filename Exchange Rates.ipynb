## Exchange Rate Analysis.
The goal of this analyis is to explore the euro to dollars exchange the exchange rate in the last decade(2011-2021). The exchange rates before and during the pandemic are also going to be explored and compared as well.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.style as style
```

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

# Cleaning The Data.

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['US_dollar', 'Time']].copy()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

# Rolling Mean

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
euro_to_dollar.tail()
```

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.title('Average Exchange Rate (Rolling Window: 30)')
plt.show()
```

```python
last_decade = euro_to_dollar[euro_to_dollar['Time'].dt.year >2010]
covid = euro_to_dollar[(euro_to_dollar['Time'].dt.year >2020)|
                     (euro_to_dollar['Time'].dt.year >2019)]

print(last_decade.tail())
```

```python
style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_ylim(0.8, 1.5)
ax.set_yticks([1.0, 1.2, 1.4, 1.6])
ax.set_yticklabels(['1.0', '1.2','1.4', '1.6'],
                   alpha=0.3)
ax.grid(alpha=0.5)
ax.tick_params(bottom=0, left=0)
for location in ['left', 'right', 'bottom', 'top']:
    ax.spines[location].set_visible(False)
        
ax.plot(last_decade['Time'], last_decade['rolling_mean'], linewidth=2, color='#A6D785')
ax.plot(covid['Time'], covid['rolling_mean'], color='#DC143C')
ax.axvspan(xmin=737420, xmax=737815, ymin=0.09, color='grey',alpha=0.3)
ax.text(733650, 1.68, 'EURO-DOLLAR RATE AVERAGED 1.4 IN THE FIRST HALF OF THE DECADE AND 1.2 IN THE SECOND HALF',
       weight='bold')
ax.text(733650, 1.63, 'EURO-DOLLAR EXCHANGE RATE IN THE LAST DECADE')
ax.text(733650, 0.6, 'Â©DATAQUEST' + ' '*200 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)
plt.show()
```
